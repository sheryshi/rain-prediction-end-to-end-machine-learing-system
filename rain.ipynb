***DROPING ROWS CONTAINING NULL VALUES***

```python
df=df.dropna()
df
```

***PREPARING A ACCURATE DATA ***

```python
df=pd.DataFrame(df)
df.columns=df.iloc[0]
```

```python
df=df.drop(df.index[0])
```

```python
df=df.reset_index()
```

```python
df
```

```python
df=df.drop('index',axis=1)
df
```

```python
df=df.rename(columns={'MO':'month','DY':'date','T2MDEW':'dew_point',	'T2M_MIN':'tmin','T2M_MAX':'tmax','RH2M':'rel_humidity','PRECTOTCORR':'preci','PS':'press_kPa','WS50M_MAX':'wind_speed'})
```

```python
df
```

```python
df.shape
```

```python
df.size
```

```python
df.info()
```

```python
df.YEAR=pd.to_datetime(df.YEAR)
df
```

```python
df['day']=df.YEAR.dt.day
df['month']=df.YEAR.dt.month
```

strip function is used to remove whitespace

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
df.dtypes
```

CHANGING THE ALL THE COLUMNS TO FLOAT DATATYPE

```python

objects=[ features for features in df.columns if df[features].dtype=='O']
objects
```

```python
for i in objects:
    
    df[i]=df[i].astype(float)
```
